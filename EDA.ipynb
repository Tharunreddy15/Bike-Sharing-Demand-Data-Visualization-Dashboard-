```python
#Q4.rename seasons
season_map={1:'spring' ,2:'summer', 3:'fall', 4:'winter'}
df['season_name']=df['season'].map(season_map)
```

```python
#Q5.total by year 
totals=df.groupby('year')[['casual','registered','count']].sum().reset_index()
print(totals)
```

```python
#Q6.mean of the hourly rental bike count by season
season_mean=df.groupby('season_name')['count'].mean().sort_values(ascending=False)
print(season_mean)
```

```python
#Q7.registered users and non-registered users 
by_work=df.groupby(['year','workingday'])[['registered']].sum()
print(by_work)
```

```python
by_work1=df.groupby(['year','workingday'])[['casual']].sum()
print(by_work1)
```

```python
#Q8.calculating highest and lowest number of rentals 
import calendar 
monthly = df.groupby(['year','month'])['count'].sum().reset_index()
monthly['month_name']=monthly['month'].apply(lambda x:calendar.month_name[x])
def highest_lowest(year):
    data=monthly[monthly['year']==year]
    max_row=data.loc[data['count'].idxmax()]
    min_row=data.loc[data['count'].idxmin()]
    print(f"\nYear {year}:")
    print("Highest rentals :", max_row['month_name'], "→", max_row['count'])
    print("Lowest rentals  :", min_row['month_name'], "→", min_row['count'])

highest_lowest(2011)
highest_lowest(2012)
```

```python
#Q9.mean of rental counts by weather
weather_map={
  1:'Clear/Partly Cloudy',
  2:'Mist/Cloudy',
  3:'Light Rain/Snow',
  4:'Heavy Precip/Fog'
}
df['weather_name']=df['weather'].map(weather_map)
weather_mean=df.groupby('weather_name')['count'].mean().sort_values()
print(weather_mean)
```

```python
#Q10.calculating the correlation
numeric_cols = ['temp','atemp','humidity','windspeed','casual','registered','count','hour']
corr = df[numeric_cols].corr()
print(corr['count'].sort_values(ascending=False))
```

```python
#Q11.creatin a categorical 
bins = [0,6,12,18,24]
labels = ['night','morning','afternoon', 'evening']
df['day_period']= pd.cut(df['hour'], bins=bins, labels=labels,right=False)
print(df[['hour','day_period']])
```

```python
#Q12. generating a pivot table by periods 
pivot = pd.pivot_table(df, values='count', index='day_period', columns='workingday', aggfunc='mean')
print(pivot)
```
